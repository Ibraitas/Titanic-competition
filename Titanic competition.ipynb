```python
X = pd.read_csv("./Data/train.csv")
X_test = pd.read_csv("./Data/test.csv")
data = pd.concat([X, X_test], ignore_index=True, sort=False)
data.head()
```

```python
data['Siblings'] = data.SibSp + data.Parch
#X_test['Siblings'] = X_test.SibSp + X_test.Parch

data['Alone'] = data.Siblings == 0
#X_test['Alone'] = X_test.Siblings == 0

data['Fare_Category'] = pd.cut(data['Fare'], bins=[0,7.90,14.45,31.28,120], labels=['Low','Mid', 'High_Mid','High'])
#X_test['Fare_Category'] = pd.cut(X_test['Fare'], bins=[0,7.90,14.45,31.28,120], labels=['Low','Mid', 'High_Mid','High'])
```

```python
data.Embarked.fillna(data.Embarked.mode()[0], inplace = True)
data.Cabin = data.Cabin.fillna('NA')

data['Salutation'] = data.Name.apply(lambda name: name.split(',')[1].split('.')[0].strip())
grp = data.groupby(['Sex', 'Pclass'])
grp.Age.apply(lambda x: x.fillna(x.median()))
data.Age.fillna(data.Age.median, inplace = True)

data['Sex'] = LabelEncoder().fit_transform(data['Sex'])
#X_test['Sex'] = LabelEncoder().fit_transform(X_test['Sex'])

data = pd.concat([data,pd.get_dummies(data.Cabin, prefix="Cabin"),
                pd.get_dummies(data.Embarked, prefix="Emb", drop_first = True),
                pd.get_dummies(data.Salutation, prefix="Title", drop_first = True),
                pd.get_dummies(data.Fare_Category, prefix="Fare", drop_first = True),
                pd.get_dummies(data.Pclass, prefix="Class", drop_first = True)], axis=1)

data.drop(['Pclass', 'Fare','Cabin', 'Fare_Category','Name','Salutation', 'Ticket','Embarked', 'SibSp', 'Parch', 'Age'], axis=1, inplace=True)
```

```python
from sklearn.model_selection import train_test_split

X_test = data[data.Survived.isnull()]
X_test = X_test.drop(['Survived'], axis = 1)

train_data = data
train_data = train_data.dropna()
y_train = train_data['Survived']
train_data  = train_data.drop(['Survived'], axis = 1)

X_train, X_valid, y_train, y_valid = train_test_split(train_data, y_train, train_size=0.8, test_size=0.2,
                                                                random_state=0)
X_train.shape
```

```python
"""model = XGBClassifier(n_estimators=1000, learning_rate=0.01, n_jobs=4)"""

model = RandomForestClassifier(criterion='entropy',
                             n_estimators=900,
                             min_samples_split=10,
                             min_samples_leaf=1,
                             max_features='auto',
                             oob_score=True,
                             random_state=1,
                             n_jobs=-1)
"""model.fit(X_train, y_train, 
          early_stopping_rounds=5, 
          eval_set=[(X_valid, y_valid)], 
          verbose=False)"""
"""print("Accuracy: "+repr(round(model.score(X_valid, y_valid) * 100, 2)) + "%")
result_rf=cross_val_score(model,X_train,y_train,cv=10,scoring='accuracy')
print('The cross validated score is:',round(result_rf.mean()*100,2))
y_pred = cross_val_predict(model,X_train,y_train,cv=10)
sns.heatmap(confusion_matrix(y_train,y_pred),annot=True,fmt='3.0f',cmap="summer")
plt.title('Confusion_matrix for RF', y=1.05, size=15)"""
```
